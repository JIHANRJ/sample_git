# Test_size VS Accuracy Graphs
In this lab, I try to optimise each ML algorithm by plotting Test_size versus Accuracy graph, then using it to predict the best accuracy and dataset size for training.

 **Note: *Not all ML algorithms are linear or straight lined*

 **test_sizes = np.arange(0.01, 0.95, 0.01)**: Creates a list of number test_sizes from 0.01 to 0.095. Values above a certain threshold causes problems for a few ML algorithms.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/ASE_Lab/heart.csv")
features = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal"]#features from heart.csv
X = df[features]
y = df['target']#lables

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.19, random_state=0)#dividing dataset
```

```python
#Decision Tree

test_sizes = np.arange(0.01, 0.95, 0.01)

test_size_values = []
accuracy_values = []

accuracy_max=0
test_size_max=0.09
for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
    dt_classifier = DecisionTreeClassifier(random_state=0)
    dt_classifier.fit(X_train, y_train)
    y_pred_dt = dt_classifier.predict(X_test)
    accuracy_dt = accuracy_score(y_test, y_pred_dt)
    test_size_values.append(test_size)
    accuracy_values.append(accuracy_dt)
    if accuracy_max<=accuracy_dt:
      accuracy_max=accuracy_dt
      test_size_max=test_size

plt.figure(figsize=(10, 6))
#plt.plot(test_size_values, accuracy_values, marker='o', linestyle='-')
plt.plot(test_size_values, accuracy_values, linestyle='-')
plt.title('Test Size vs. Accuracy')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
print("Best Accuracy found for Test Size = ",test_size_max)
print("Best Accuracy = ",accuracy_max)
```

```python
#Random Forest
test_sizes = np.arange(0.01, 0.95, 0.01)

test_size_values_rf = []
accuracy_values_rf = []

accuracy_max=0
test_size_max=0.01
for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
    rf_classifier = RandomForestClassifier(random_state=0)
    rf_classifier.fit(X_train, y_train)
    y_pred_rf = rf_classifier.predict(X_test)
    accuracy_rf = accuracy_score(y_test, y_pred_rf)
    test_size_values_rf.append(test_size)
    accuracy_values_rf.append(accuracy_rf)
    if accuracy_max<=accuracy_rf:
      accuracy_max=accuracy_rf
      test_size_max=test_size


plt.figure(figsize=(10, 6))
plt.plot(test_size_values_rf, accuracy_values_rf, linestyle='-')
plt.title('Test Size vs. Accuracy (Random Forest)')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
print("Best Accuracy found for Test Size = ",test_size_max)#note this will only give the largest test size.
print("Best Accuracy = ",accuracy_max)
```

```python
#SVM
test_sizes = np.arange(0.01, 0.95, 0.01)

test_size_values_svm = []
accuracy_values_svm = []

accuracy_max=0
test_size_max=0.01
for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
    svm_classifier = SVC(kernel='linear', C=1.0)
    svm_classifier.fit(X_train, y_train)
    y_pred_svm = svm_classifier.predict(X_test)
    accuracy_svm = accuracy_score(y_test, y_pred_svm)
    test_size_values_svm.append(test_size)
    accuracy_values_svm.append(accuracy_svm)
    if accuracy_max<=accuracy_svm:
      accuracy_max=accuracy_svm
      test_size_max=test_size


plt.figure(figsize=(10, 6))
plt.plot(test_size_values_svm, accuracy_values_svm, linestyle='-')
plt.title('Test Size vs. Accuracy ')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
print("Best Accuracy found for Test Size = ",test_size_max)#note this will only give the largest test size.
print("Best Accuracy = ",accuracy_max)
```

```python
#Logistic Regression
test_sizes = np.arange(0.01, 0.95, 0.01)

test_size_values_lr = []
accuracy_values_lr = []

accuracy_max_lr = 0
test_size_max_lr = 0.01

for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
    lr_classifier = LogisticRegression(max_iter=2000, random_state=0)
    lr_classifier.fit(X_train, y_train)
    y_pred_lr = lr_classifier.predict(X_test)
    accuracy_lr = accuracy_score(y_test, y_pred_lr)
    test_size_values_lr.append(test_size)
    accuracy_values_lr.append(accuracy_lr)
    if accuracy_lr >= accuracy_max_lr:
        accuracy_max_lr = accuracy_lr
        test_size_max_lr = test_size

plt.figure(figsize=(10, 6))
plt.plot(test_size_values_lr, accuracy_values_lr, linestyle='-')
plt.title('Test Size vs. Accuracy ')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

print("Test Size =", test_size_max_lr)
print("Best Accuracy =", accuracy_max_lr)
```

```python
#KNN
test_sizes = np.arange(0.01, 0.95, 0.01)

test_size_values_knn = []
accuracy_values_knn = []

accuracy_max_knn = 0
test_size_max_knn = 0.01

for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
    knn_classifier = KNeighborsClassifier(n_neighbors=5)
    knn_classifier.fit(X_train, y_train)
    y_pred_knn = knn_classifier.predict(X_test)
    accuracy_knn = accuracy_score(y_test, y_pred_knn)
    test_size_values_knn.append(test_size)
    accuracy_values_knn.append(accuracy_knn)
    if accuracy_knn >= accuracy_max_knn:
        accuracy_max_knn = accuracy_knn
        test_size_max_knn = test_size

plt.figure(figsize=(10, 6))
plt.plot(test_size_values_knn, accuracy_values_knn, linestyle='-')
plt.title('Test Size vs. Accuracy')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

print("Test Size =", test_size_max_knn)
print("Best Accuracy =", accuracy_max_knn)
```

```python
#Naive Bayes
test_sizes = np.arange(0.09, 0.91, 0.01)

test_size_values_nb = []
accuracy_values_nb = []

accuracy_max_nb = 0
test_size_max_nb = 0.09

for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
    nb_classifier = GaussianNB()
    nb_classifier.fit(X_train, y_train)
    y_pred_nb = nb_classifier.predict(X_test)
    accuracy_nb = accuracy_score(y_test, y_pred_nb)
    test_size_values_nb.append(test_size)
    accuracy_values_nb.append(accuracy_nb)
    if accuracy_nb >= accuracy_max_nb:
        accuracy_max_nb = accuracy_nb
        test_size_max_nb = test_size

plt.figure(figsize=(10, 6))
plt.plot(test_size_values_nb, accuracy_values_nb, linestyle='-')
plt.title('Test Size vs. Accuracy ')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

print("Test Size =", test_size_max_nb)
print("Best Accuracy =", accuracy_max_nb)
```
